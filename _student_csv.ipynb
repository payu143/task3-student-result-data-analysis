```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df["Total"] = (
df["Mathematics"] +
df["Communication System"] + 
df["Java Programming"] + 
df["Python Programming"] +
df["Basic Electronics"]
)
df["percentage"] = (df["Total"] /500)*100
df.head()
```

```python
df.columns
```

```python
grades = []

for i in df["percentage"]:
    if i >= 75:
        grades.append("Distinction")
    elif i >= 60:
        grades.append("First Class")
    elif i >= 50:
        grades.append("Second Class")
    elif i >= 40:
        grades.append("Pass")
    else:
        grades.append("Fail")

df["Grade"] = grades
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df,x="Grade",
palette="Set2")
plt.title("Grade Distribution")
plt.xlabel("Grades")
plt.ylabel("Number of students")
plt.show()
```
